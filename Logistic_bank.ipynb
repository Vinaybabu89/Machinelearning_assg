```python
from sklearn.preprocessing import LabelEncoder
enc=LabelEncoder()
bank['y']=enc.fit_transform(bank['y'])
```

```python
bank.columns
```

```python
bank_dum=pd.get_dummies(bank[[ 'job', 'marital', 'education', 'default','housing',
       'loan', 'contact', 'month','poutcome']])
bank_dum
```

```python
bank.drop([ 'job', 'marital', 'education', 'default','housing',
       'loan', 'contact', 'month','poutcome'],inplace=True,axis=1)
```

```python
final_df=pd.concat([bank_dum,bank],axis=1)
final_df
```

```python
X=final_df.iloc[:,0:51]
Y=final_df.iloc[:,51]
Y
```

```python
from sklearn.linear_model import LogisticRegression
classifier=LogisticRegression()
classifier.fit(X,Y)
```

```python
classifier.coef_
```

```python
proba=pd.DataFrame(classifier.predict_proba(X))
proba
```

```python
pred=classifier.predict(X)
len(pred)
```

```python
New_df=pd.DataFrame({'Actual':bank['y'],'Predicted':pred})
New_df=pd.concat([New_df,proba],axis=1)
New_df
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix = confusion_matrix(Y,pred)
print (confusion_matrix)
```

```python
from sklearn import metrics
print("Accuracy:",metrics.accuracy_score(Y, pred))
print("Precision:",metrics.precision_score(Y, pred))
print("Recall:",metrics.recall_score(Y, pred))
```

```python
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score

fpr, tpr, thresholds = roc_curve(Y, pred)

auc = roc_auc_score(Y, pred)

import matplotlib.pyplot as plt
plt.plot(fpr, tpr, color='orange', label='ROC')
```
